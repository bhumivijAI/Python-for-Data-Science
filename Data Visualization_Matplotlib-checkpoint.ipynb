# Matplotlib

Matplotlib is a comprehensive library for creating static,animated,and interactive visualizations in python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

A "figure " means the entire image containing all the subplots."Axes" refers to the individual subplots.

```python
fig,ax = plt.subplots()
print(fig,ax)
```

```python
fig,ax = plt.subplots(6)
print(fig,ax)
```

```python
# You can change the figure size 
fig,ax = plt.subplots(figsize=(2,5))
```

```python
# You can create many axes(plots) in a single figure
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1,figsize=(5,5))
```

```python
print(fig,ax1)
```

 # Bar Chart

```python
# lets create a bar graph
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
ax.bar(x,y)
```

```python
help(ax.bar)
```

```python
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
ax.bar(x,y)
ax.set_xlabel("Product")
ax.set_ylabel("Units Sold")
ax.set_title("Units sold for each product")
ax.grid(True)  # Shows grid
```

```python
# Lets change the color too
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
color_bar = ["lightblue","lavender","pink"]
ax.bar(x,y,color=color_bar)
```
